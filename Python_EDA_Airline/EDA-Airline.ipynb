```python
import matplotlib.pyplot as plt
import seaborn as sns
def grafikler(data):
    numeric_columns= [col for col in data.columns if data[col].dtype != "O"] #sayısal değerleri alır
    for i in numeric_columns:
        fig, axes = plt.subplots(nrows=1,ncols=3, figsize=(20,4))
        sns.histplot(x=data[i],bins=10,ax=axes[0]) # hangi sütunu çizdirecek i ile veriyor
        axes[0].set_title(i)

        sns.boxplot(x=data[i],ax=axes[1])
        axes[1].set_title(i)

        sns.kdeplot(x=data[i],ax=axes[2])
        axes[2].set_title(i)
        plt.show()
```

```python
grafikler(data)
```

```python
memnuniyet_haritasi = {
    'neutral or dissatisfied': 0,
    'satisfied': 1
}
data['Num_memuniyet']= data['Memnuniyet'].map(memnuniyet_haritasi)
```

```python
import warnings
warnings.filterwarnings('ignore')

def kategorik_grafikler(data):
    categorical_columns = [col for col in data.columns if data[col].dtype == "O"]

    for i in categorical_columns:
        # Sıklık tablosu (bar plot)
        fig, axes = plt.subplots(nrows=1,ncols=3, figsize=(20,4))
        sns.countplot(data=data, x=i,ax=axes[0])
        plt.xticks(rotation=90)
        axes[0].set_title(f'{i} Sayım Tablosu')

        # Pasta grafiği (pie chart)
        data[i].value_counts().plot.pie(autopct='%1.1f%%',ax=axes[1])
        plt.ylabel('')
        axes[1].set_title(f'{i} Pasta Grafiği')

        # Nokta grafiği (point plot)
        sns.pointplot(data=data, x=i, y='Num_memuniyet', ci=None,ax=axes[2])
        plt.xticks(rotation=90)
        axes[2].set_title(f'{i} Nokta Grafiği')

# Örnek kullanım
kategorik_grafikler(data)
```

```python
corr_matrix = data.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True) # annot=True: değerleri yazar
plt.show()
```
